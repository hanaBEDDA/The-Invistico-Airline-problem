```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def check_linearity(dataframe, dependent_variable, independent_variables):
    """
    Check linearity in the data using scatter plots and correlation coefficients.

    Parameters:
        dataframe (pandas.DataFrame): The DataFrame containing the data.
        dependent_variable (str): The name of the dependent variable.
        independent_variables (list): A list of names of independent variables.

    Returns:
        None
    """
    # Check if the specified variables exist in the DataFrame
    if dependent_variable not in dataframe.columns:
        raise ValueError(f"Dependent variable '{dependent_variable}' not found in the DataFrame.")
    for variable in independent_variables:
        if variable not in dataframe.columns:
            raise ValueError(f"Independent variable '{variable}' not found in the DataFrame.")

    # Scatter plot each independent variable against the dependent variable
    for independent_variable in independent_variables:
        plt.figure(figsize=(8, 6))
        plt.scatter(dataframe[independent_variable], dataframe[dependent_variable], alpha=0.5)
        plt.xlabel(independent_variable)
        plt.ylabel(dependent_variable)
        plt.title(f"{dependent_variable} vs {independent_variable}")
        plt.show()

        # Calculate the correlation coefficient
        correlation_coefficient = np.corrcoef(dataframe[independent_variable], dataframe[dependent_variable])[0, 1]
        print(f"The correlation coefficient between {dependent_variable} and {independent_variable} is: {correlation_coefficient:.4f}")

# Example usage:
# Suppose you have a DataFrame called 'data' with columns 'dependent_var', 'independent_var1', and 'independent_var2'
# You can call the function like this:
check_linearity(Invistico_Airline, 'satisfaction', ['leg_room_service', 'baggage_handling'])
```
